```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
import pandas as pd
import dash_table

df = pd.read_excel(
    "https://github.com/chris1610/pbpython/blob/master/data/salesfunnel.xlsx?raw=True"
)
mgr_options = df["Manager"].unique()
product_options=df["Product"].unique()
fnameDict=df.groupby('Rep')['Product'].apply(list).to_dict()


app = dash.Dash()

app.layout = html.Div([
    html.H2("Sales Funnel Report"),
    html.Div(
        [
            dcc.Dropdown(
                id="Manager",
                options=[{
                    'label': i,
                    'value': i
                } for i in mgr_options],
                multi=True,
                value='All Manager'),
        ],
        style={'width': '25%',
               'display': 'inline-block','float':'left'}),

        html.Div(
        [
            dcc.Dropdown(
                id="product",
                options=[{
                    'label': i,
                    'value': i
                } for i in product_options],
                multi=True,
                value='All product'),
        ],
        style={'width': '25%',
               'display': 'inline-block','float':'left','margin-bottom':'200px'}),
    
    
    #Graph styling
        html.Div([
            dcc.Graph(id='funnel-graph')],
    style={'width': '50%','margin-left':'-750px','margin-top':'50px', 'display': 'inline-block'}),
         html.Div([
            dcc.Graph(id='funnel-graph2')],
    style={'width': '50%','margin-left':'30px','margin-top':'50px', 'display': 'inline-block'})
    ])

@app.callback(dash.dependencies.Output('funnel-graph', 'figure'),
    [dash.dependencies.Input('Manager', 'value'),dash.dependencies.Input('product', 'value')])
def update_graph(Manager,product):
    if Manager=='All Manager':
        df1=df.copy()
    else:
        df1 = df[(df['Manager'].isin(Manager))]
    if product=='All product':
        df2=df1.copy()
    else:
        df2 = df1[(df1['Product'].isin(product))] 
     
    
    pv = pd.pivot_table(
        df2,
        index=['Name'],
        values=['Quantity'],
        aggfunc=sum,
        fill_value=0)

    trace1 = go.Bar(x=pv.index, y=pv[('Quantity')])


    return {
        'data': [trace1],
        'layout':
        go.Layout(
            title='Customer Order Status for {} {}'.format(Manager,product),
            barmode='stack')
    }

@app.callback(dash.dependencies.Output('funnel-graph2', 'figure'),
    [dash.dependencies.Input('Manager', 'value'),dash.dependencies.Input('product', 'value')])
def update_graph(Manager,product):
    if Manager=='All Manager':
        df1=df.copy()
    else:
        df1 = df[(df['Manager'].isin(Manager))]
    if product=='All product':
        df2=df1.copy()
    else:
        df2 = df1[(df1['Product'].isin(product))]   

    
    pv = pd.pivot_table(
        df2,
        index=['Name'],
        values=['Price'],
        aggfunc=sum,
        fill_value=0)

    trace1 = go.Bar(x=pv.index, y=pv[('Price')], name='price')


    return {
        'data': [trace1],
        'layout':
        go.Layout(
            title='Customer Order Status for {} {}'.format(Manager,product),
            barmode='stack')
    }

app.css.append_css({
    'external_url': 'https://codepen.io/chriddyp/pen/bWLwgP.css'
})

if _name_ == '_main_':
    app.run_server(debug=False,port=5696)
```
