```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
import pandas as pd
import dash_table

df = pd.read_excel(
    "https://github.com/chris1610/pbpython/blob/master/data/salesfunnel.xlsx?raw=True"
)
mgr_options = df["Manager"].unique()
product_options=df["Product"].unique()



app = dash.Dash()

app.layout = html.Div([
    html.H2("Sales Funnel Report"),
    html.Div(
        [
            dcc.Dropdown(
                id="Manager",
                options=[{
                    'label': i,
                    'value': i
                } for i in mgr_options],
                multi=True,
                value='Debra Henley'),
        ],
        style={'width': '25%',
               'display': 'inline-block'}),

        html.Div(
        [
            dcc.Dropdown(
                id="product",
                options=[{
                    'label': i,
                    'value': i
                } for i in product_options],
                multi=True,
                value='Maintenance'),
        ],
        style={'width': '25%',
               'display': 'inline-block'}),
    dcc.Graph(id='funnel-graph'),
    dcc.Graph(id='funnel-graph2')
])


@app.callback(dash.dependencies.Output('funnel-graph', 'figure'),
    [dash.dependencies.Input('Manager', 'value'),dash.dependencies.Input('product', 'value')])
def update_graph(Manager,product):
    new_df = df[(df['Manager'].isin(Manager))]
    df1 = new_df[(new_df['Product'].isin(product))] 
     
    
    pv = pd.pivot_table(
        df1,
        index=['Name'],
        values=['Quantity'],
        aggfunc=sum,
        fill_value=0)
    pv["pct"]=pv["Quantity"]/pv["Quantity"].sum()
    trace1 = go.Bar(x=pv.index, y=pv[('Quantity')], name='Quantity')
    trace2=go.Scatter(x=pv.index,y=pv[('pct')],name='pct')

    layout = go.Layout(
    title='Quantity',
    yaxis=dict(
        title='Quantity'
    ),
    yaxis2=dict(
        title='pct',
        overlaying='y',
        side='right'
    )
)
    data=[trace1,trace2]


    return go.Figure(
        data=data,
        layout=layout)


if __name__ == '__main__':
    app.run_server(debug=False)
```
